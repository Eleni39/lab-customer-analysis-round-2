```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
X=data.drop('total claim amount', axis=1)
y=data ['total claim amount']
```

```python
X.head()
```

```python
X_cat = X.select_dtypes('object')
```

```python
X_num = X.select_dtypes('number')
```

```python
X_cat
```

```python
X_num
```

```python
from sklearn.preprocessing import MinMaxScaler
MinMaxtransformer = MinMaxScaler().fit(X_num)
x_normalized = MinMaxtransformer.transform(X_num)
print(x_normalized.shape)
x_normalized = pd.DataFrame(x_normalized,columns=X_num.columns)
x_normalized.head()
```

```python
x_normalized.describe().T
```

```python
X_cat.head()
```

```python
X_cat = X_cat.drop('customer', axis =1)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(X_cat)
cols = encoder.get_feature_names(input_features=X_cat.columns)
X_cat_encode = pd.DataFrame(encoder.transform(X_cat).toarray(),columns=cols)
X_cat_encode.head()
```

```python
X = pd.concat([x_normalized, X_cat_encode], axis = 1)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Here is my linear regression code. 
from sklearn.linear_model import LinearRegression as LinReg

lm=LinReg()    # model
lm.fit(X_train, y_train)   # model training
y_pred_lm=lm.predict(X_test)   # model prediction

#lm = linear_model.LinearRegression()
#lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_lm)
```

```python
predictions = lm.predict(X_test)
r2_score(y_test, predictions)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,predictions))
```
